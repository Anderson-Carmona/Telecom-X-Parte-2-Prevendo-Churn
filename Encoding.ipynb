```python
Descrição
Transforme as variáveis categóricas em formato numérico para torná-las compatíveis com algoritmos de machine learning. Utilize um método de codificação adequado, como o one-hot encoding.

Transformar variáveis categóricas em numéricas é um passo essencial no pré-processamento de dados, pois a maioria dos algoritmos de Machine Learning (como regressão linear, SVM e redes neurais) baseia-se em cálculos matemáticos que exigem entradas numéricas.
Aqui está uma explicação prática de como aplicar o One-Hot Encoding e quando ele é a melhor escolha.
________________________________________
O que é One-Hot Encoding?
O One-Hot Encoding cria novas colunas binárias (0 ou 1) para cada categoria única de uma variável. Isso evita o problema da "ordem implícita": se codificássemos "Verde" como 1, "Azul" como 2 e "Vermelho" como 3, o modelo poderia erroneamente pensar que Vermelho é "maior" ou "melhor" que Verde.
Exemplo Prático com Python (Pandas)
Suponha que temos um conjunto de dados de frutas com as cores: Vermelho, Verde e Azul.
Python
import pandas as pd

# Criando um DataFrame de exemplo
data = {'Fruta': ['Maçã', 'Melancia', 'Mirtilo', 'Morango'],
        'Cor': ['Vermelho', 'Verde', 'Azul', 'Vermelho']}
df = pd.DataFrame(data)

# Aplicando One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=['Cor'])

print(df_encoded)
Resultado esperado:
Fruta	Cor_Azul	Cor_Verde	Cor_Vermelho
Maçã	0	0	1
Melancia	0	1	0
Mirtilo	1	0	0
Morango	0	0	1
________________________________________
Qual método escolher?
A escolha do método depende da natureza dos seus dados:
1.	One-Hot Encoding: Ideal para variáveis nominais (sem ordem lógica), como Cores, Cidades ou Marcas.
o	Ponto de atenção: Se a variável tiver centenas de categorias, isso pode criar colunas demais (a "maldição da dimensionalidade").
2.	Ordinal Encoding: Ideal para variáveis ordinais (com ordem lógica), como Nível de Escolaridade (Fundamental < Médio < Superior) ou Tamanho (P < M < G).
o	Aqui, atribuímos números sequenciais (0, 1, 2...) que preservam a hierarquia.
________________________________________
Boas Práticas
•	Dummy Variable Trap: Ao usar One-Hot Encoding para modelos de regressão, é comum remover uma das colunas geradas (drop_first=True) para evitar a multicolinearidade perfeita.
•	Scikit-Learn: Para pipelines de produção, prefira o OneHotEncoder da biblioteca Scikit-Learn, pois ele facilita a aplicação dos mesmos critérios nos dados de teste.
```
